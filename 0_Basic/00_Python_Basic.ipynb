## Python Basic

### 基本運算:
```python
# 四則運算
>>> 2 * 3 + 10 / 5 - 2**2
4.0

# import 數學套件，sqrt為開根號 (square root)
>>> from math import sqrt
>>> (15-5) / sqrt(4)
5.0

# 取餘數
>>> 5 % 2
1

```
